##***EJERCICIO 1***

Simular el lanzamiento de un dado equilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de lanzamiento de dado
def lanzar_dado():
    return np.random.randint(1, 7)  # El límite superior es exclusivo

# Número de lanzamientos
num_lanzamientos = 1000

# Realizar los lanzamientos
resultados = np.array([lanzar_dado() for _ in range(num_lanzamientos)])

# Contar resultados
contador_2 = np.sum(resultados == 2)
contador_3 = np.sum(resultados == 3)
contador_5 = np.sum(resultados == 5)

# Calcular probabilidades aproximadas
probabilidad_2 = contador_2 / num_lanzamientos
probabilidad_3 = contador_3 / num_lanzamientos
probabilidad_5 = contador_5 / num_lanzamientos

# Imprimir probabilidades
print("Probabilidad de obtener el número 2:", probabilidad_2)
print("Probabilidad de obtener el número 3:", probabilidad_3)
print("Probabilidad de obtener el número 5:", probabilidad_5)

# Graficar histograma
resultados_posibles = np.array([2, 3, 5])
probabilidades = np.array([probabilidad_2, probabilidad_3, probabilidad_5])

plt.bar(resultados_posibles, probabilidades)
plt.xlabel('Resultados')
plt.ylabel('Probabilidades')
plt.title('Histograma de probabilidades de obtener 2, 3 y 5 en un dado de 6 caras')
plt.xticks(resultados_posibles)
plt.show()
```

```python
#punto 5

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Solicitar al usuario la cantidad N de datos
N = int(input("Ingrese la cantidad de datos a cargar: "))

# Generar datos aleatorios
np.random.seed(42)  # Para reproducibilidad
sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], size=N)
edad = np.random.randint(28, 61, size=N)
altura = np.random.uniform(165.0, 210.0, size=N)
peso = np.random.uniform(50, 110, size=N)
horas_descanso = np.random.uniform(4, 10, size=N)
horas_trabajo_estudio = np.random.uniform(2, 14, size=N)
horas_ejercicio = np.random.uniform(0.5, 3, size=N)

# Crear un DataFrame
data = pd.DataFrame({
    'SEXO': sexo,
    'EDAD': edad,
    'ALTURA': altura,
    'PESO': peso,
    'HORAS_DESCANSO': horas_descanso,
    'HORAS_TRABAJO_ESTUDIO': horas_trabajo_estudio,
    'HORAS_EJERCICIO': horas_ejercicio
})

# Mostrar los primeros datos en forma de tabla
print(data.head())

# Calcular la media de los datos
media_altura = np.mean(altura)
media_peso = np.mean(peso)
media_horas_descanso = np.mean(horas_descanso)
media_horas_trabajo_estudio = np.mean(horas_trabajo_estudio)
media_horas_ejercicio = np.mean(horas_ejercicio)

print("Media de Altura:", media_altura)
print("Media de Peso:", media_peso)
print("Media de Horas de Descanso:", media_horas_descanso)
print("Media de Horas de Trabajo/Estudio:", media_horas_trabajo_estudio)
print("Media de Horas de Ejercicio:", media_horas_ejercicio)


# Graficar los datos de peso y altura
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(peso, bins=20, color='blue', alpha=0.7)
plt.title('Distribución de Pesos')
plt.xlabel('Peso')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
plt.hist(altura, bins=20, color='green', alpha=0.7)
plt.title('Distribución de Alturas')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()


# Graficar correlación entre peso y altura
plt.scatter(altura, peso, alpha=0.5)
plt.title('Correlación entre Peso y Altura')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.show()

# Graficar las horas de descanso, trabajo/estudio y ejercicio
plt.plot(horas_descanso, label='Horas de Descanso', marker='o')
plt.plot(horas_trabajo_estudio, label='Horas de Trabajo/Estudio', marker='o')
plt.plot(horas_ejercicio, label='Horas de Ejercicio', marker='o')
plt.title('Horas de Descanso, Trabajo/Estudio y Ejercicio')
plt.xlabel('Persona')
plt.ylabel('Horas')
plt.legend()
plt.show()
```

```python
#se define una función para generar y limpiar los outliers
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Llamamos a la función para cada columna numérica donde desees detectar outliers
columns_to_check = ["Age"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df, col)
    outliers_list.append(outliers)

# Concatenamos todos los outliers encontrados en un solo DataFrame
outliers_df = pd.concat(outliers_list)

# Eliminamos los outliers del conjunto de datos original
df_cleaned1 = df.drop(outliers_df.index)

# Imprimimos la cantidad de outliers detectados
print("Número de outliers detectados:", len(outliers_df))
print(outliers)
```
