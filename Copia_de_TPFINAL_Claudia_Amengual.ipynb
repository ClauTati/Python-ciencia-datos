```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/StudentsPerformance.csv')

# Gráfico de histogramas
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.histplot(data['math score'], kde=True)
plt.title('Histograma - Math Score')

plt.subplot(1, 3, 2)
sns.histplot(data['reading score'], kde=True)
plt.title('Histograma - Reading Score')

plt.subplot(1, 3, 3)
sns.histplot(data['writing score'], kde=True)
plt.title('Histograma - Writing Score')

plt.tight_layout()
plt.show()

# Gráfico de boxplots
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.boxplot(data=data[['math score']])
plt.title('Boxplot - Math Score')

plt.subplot(1, 3, 2)
sns.boxplot(data=data[['reading score']])
plt.title('Boxplot - Reading Score')

plt.subplot(1, 3, 3)
sns.boxplot(data=data[['writing score']])
plt.title('Boxplot - Writing Score')

plt.tight_layout()
plt.show()
```

Histogramas:

Las puntuaciones de math score parecen estar distribuidas de manera más o menos normal, con una concentración alrededor de los valores centrales.
Las puntuaciones de reading score también muestran una distribución similar a la normal, con una concentración en los valores centrales.
Las puntuaciones de writing score también siguen una distribución cercana a la normal, aunque con algunos valores más bajos.

Boxplots:

En el boxplot de math score, hay algunos valores atípicos por debajo del límite inferior, pero en general, las puntuaciones están dentro del rango intercuartil (IQR).
En el boxplot de reading score, no hay valores atípicos y las puntuaciones están bastante distribuidas dentro del IQR.
En el boxplot de writing score, también hay algunos valores atípicos por debajo del límite inferior, pero en general, las puntuaciones están dentro del IQR.
En general, las puntuaciones de las tres asignaturas parecen estar distribuidas de manera similar, con algunas diferencias en los valores atípicos y la dispersión. Las puntuaciones están centradas alrededor de valores intermedios, y hay una variedad de puntuaciones tanto en el rango bajo como en el rango alto. Estos gráficos proporcionan una visión general de la distribución y dispersión de las puntuaciones en el conjunto de datos "StudentsPerformance.csv".


```python
#Ejercicio 6

# Cargar tus datos en un DataFrame (data)
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/StudentsPerformance.csv')

# Calcular la matriz de correlación
correlation_matrix = data.corr()

# Crear un mapa de calor de la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/StudentsPerformance.csv')

# Seleccionar solo las columnas numéricas para el análisis de correlación
numeric_columns = ['math score', 'reading score', 'writing score']

# Calcular la matriz de correlación
correlation_matrix = data[numeric_columns].corr()

# Crear un mapa de calor de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlación')
plt.show()
```

En el contexto de las puntuaciones (math score, reading score, writing score):

La correlación entre math score y reading score es alta, lo que indica que los estudiantes que obtienen altas puntuaciones en matemáticas también tienden a obtener altas puntuaciones en lectura.
La correlación entre math score y writing score también es alta, lo que sugiere que los estudiantes que tienen un buen rendimiento en matemáticas tienden a tener un buen rendimiento en escritura.
La correlación entre reading score y writing score también es alta, lo que indica que los estudiantes que obtienen altas puntuaciones en lectura tienden a obtener altas puntuaciones en escritura.
En resumen, las tres puntuaciones están correlacionadas positivamente entre sí, lo que sugiere que los estudiantes que se desempeñan bien en una asignatura tienden a desempeñarse bien en las otras dos asignaturas. Esta información puede ser útil para identificar patrones de desempeño de los estudiantes y comprender las relaciones entre las asignaturas.
